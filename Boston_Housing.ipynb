### Description about the columns name
* CRIM - per capita crime rate by town
* ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
* INDUS - proportion of non-retail business acres per town.
* CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
* NOX - nitric oxides concentration (parts per 10 million)
* RM - average number of rooms per dwelling
* AGE - proportion of owner-occupied units built prior to 1940
* DIS - weighted distances to five Boston employment centres
* RAD - index of accessibility to radial highways
* TAX - full-value property-tax rate per dollor 10,000
* PTRATIO - pupil-teacher ratio by town
* B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
* LSTAT - % lower status of the population
* MEDV - Median value of owner-occupied homes in $1000's

** MEDV is the target column

### Outliers handling

```python
column = ['CRIM', 'ZN', 'B','RM','AGE','PTRATIO','MEDV']

for col in column:
    q1 = np.percentile(df[col], 25)
    q3 = np.percentile(df[col], 75)
    IQR = q3 - q1
    upper_bound = q3 + 1.5*IQR
    lower_bound = q1 - 1.5*IQR
    df[col] = np.where(df[col]>upper_bound, upper_bound, df[col])
    df[col] = np.where(df[col]<lower_bound, lower_bound, df[col])
    
```

### BoxPlot after outlier handling

```python
plt.figure(figsize = (12, 6))
number = 1
for col in column:
    if number <= 7:
        ax = plt.subplot(2, 4, number)
        sns.boxplot(df[col])
        plt.xlabel(col, fontsize = 15)
        
    number += 1
    
plt.tight_layout()
plt.show()
```

### Split the dataset to train the model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
import warnings
from sklearn.exceptions import ConvergenceWarning

warnings.filterwarnings('ignore', category=ConvergenceWarning)
```

### Importing different models

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
```

### Defining function to predict the model and check the performance

```python
def train_test(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculate the metrics
    r2 = r2_score(y_test, y_pred)
    adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
    print(f"For {model}")
    print()
    print("Train score : ", model.score(X_train, y_train))
    print("Test score : ", model.score(X_test, y_test))
    print("r2 score: ", r2)
    print("Adjusted_r2_score: ",adjusted_r2)
    print("MAE: ", mean_absolute_error(y_test, y_pred))
    print("MSE: ", mean_squared_error(y_test, y_pred))
    
    # Visualise the model perfomance using plot
    plt.figure(figsize = (10,5))
    sns.regplot(x=y_test,y=y_pred, line_kws=dict(color="r"))
    plt.title(f'{model}', fontsize = 20)
    print("-"*100)
    return
```

### Creating model to train the dataset

```python
m1 = LinearRegression()
m2 = KNeighborsRegressor(n_neighbors=5)
m3 = RandomForestRegressor()
m4 = AdaBoostRegressor()
m5 = GradientBoostingRegressor()
m6 = SVR()
```

### Calling each model for training and prediction

```python
models = [m1, m2, m3, m4, m5, m6]

for model in models:
    train_test(model, X_train, X_test, y_train, y_test)
```

* Model is getting overfit for the algorithms like KNN, Random Forest, Gradient Boosting, SVM because the differece between train score and test score is 10%

### Model coefficients interpretation

```python
# For linear regression model
coefficients = m1.coef_
intercept = m1.intercept_

# feature names in a list
feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'TAX', 'PTRATIO', 'B', 'LSTAT']

# Combine feature names with their coefficients
feature_importance = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})

# Display the coefficients
print(feature_importance)

# Interpreting the coefficients
print("Intercept:", intercept)
for feature, coef in zip(feature_names, coefficients):
    print(f"The feature '{feature}' has a coefficient of {coef:.4f}, meaning that for a one unit increase in '{feature}', the housing price changes by {coef:.4f} units, holding all other features constant.")
```

```python
# For AdaBoost regression model
feature_importance = m4.feature_importances_

# feature names in a list
feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'TAX', 'PTRATIO', 'B', 'LSTAT']

# Combine feature names with their importances
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importance})

# Sort by importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Display the feature importances
print(feature_importance_df)

# Interpreting the feature importances
for feature, importance in zip(feature_importance_df['Feature'], feature_importance_df['Importance']):
    print(f"The feature '{feature}' has an importance score of {importance:.4f}, indicating its relative importance in predicting housing prices.")
```
